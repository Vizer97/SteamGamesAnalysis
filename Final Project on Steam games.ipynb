## Cleansing the data
From the above analysis we have found that there are few missing data and to cleanse the data we are using the following method.
1) **Mode** is used to fill the **categorical data** <br>
2) **Mean** is used to fill the **numerical data**

```python
#Cleansing the data
print(games_df['Developers'].mode()[0])
print(games_df['Name'].mode()[0])
print(games_df['Publishers'].mode()[0])
print(games_df['Categories'].mode()[0])
print(games_df['Recommendations'].mode()[0])
print(games_df['Notes'].mode()[0])
print(games_df['About the game'].mode()[0])
print(games_df['Screenshots'].mode()[0])
print(games_df['Metacritic score'].mode()[0])
print(games_df['Genres'].mode()[0])
print(games_df['Header image'].mode()[0])
print(games_df['Website'].mode()[0])
print(games_df['Metacritic Rating'].mean())
print(games_df['Required age'].mean())
```

```python
#For filling categorical data
cat = ['Developers','Publishers','Website','Header image','Support url','Categories','Recommendations','Notes','About the game','Name','Screenshots','Genres','Metacritic score']
for column in cat:
    games_df[column].fillna(games_df[column].mode()[0], inplace=True)
```

```python
#For filling numerical data
num = ['Required age']

for column in num:
    games_df[column].fillna(round(games_df[column].mean(),0), inplace=True)
```

```python
#Check the cleansed data
games_df.isna().sum()
```

# 1) Hypothesis:- Whether the maximum amount spent on steam games belong to the age group above 10
### Bar Graph
To prove this hypothesis we are comparing two variables **Price** and **Required age** <br>
The best way to visualize the data is to compare it in a <b>bar graph</b>.<br>
The <b>required age</b> is taken in <b>X-axis</b> and the <b>price</b> is taken in <b>Y-axis</b>.
<br><br>

<b>Insights</b>
* From the first graph it is visible that the maximum average price spent is 17.5 dollars by individuals who belong to age 10 and the average price spent is 8 dollars by the individuals under age 10.
* From the second graph we can see that the maximum average price spent by an individual is 25 dollars and average price spent is 15 dollars by the individuals over the age of 10. 
* <b> Thus we can conclude that the maximum amount spent on steam games belong to the age group above 10.
    

```python
#Creating a sample dataframe
df = games_df[0:40000]

#Plotting a bar graph between Required age and Price for individuals under 10
data_req_1 = df.loc[((df["Required age"] >=1) & (df["Required age"] <= 10))  &  ((df["Price"] >1 ) & (df["Price"] <50 ))]
plt_1 = data_req_1.groupby(['Required age']).mean().plot(kind='bar',color=["b"], y='Price', title = 'Average amount of money spent on games online by individuals under 10')

#Plotting a bar graph between Required age and Price for individuals over 10
data_req_2 = df.loc[((df["Required age"] >=10)) & ((df["Price"] >1 ) & (df["Price"] <50 ))]
plt_2 = data_req_2.groupby(['Required age']).mean().plot(kind='bar',color=["g"], y='Price', title = 'Average amount of money spent on games online by individuals over 10')
```

# 3) Hypothesis :-  Whether Amazon Games produces high-priced games

## Box Plot
To analyze this Hypothesis we have taken Box plot representation.
We have taken the top 4 game producers and compare the prices of their games. 

**Insights**
1) From the below charts we can see where the quartiles lie with respect to price. We can clearly compare the highest quartile for all the 4 developers and see that it is high for Amazon Games. <br>
2) To confirm the confidence level of the median, notches in the first graph is used as an additional evidence to prove the hypothesis. <br>
3) We can also see that Ubisoft produces costly games than Amazon games but in very few numbers. So they are considered as outliers and it does not impact in rejecting our hypothesis.

**Thus from the above insights we can prove that Amazon Games produces high priced games compared to competitors.**


```python
#Taking the dataframe with required fields -  Developers and Price
df = games_df[['Developers','Price']]

#We take the top 4 Developers list
mylist = ['Sony', 'Ubisoft', 'Big Fish Games', 'Amazon Games']

#Filter the data of the top 4 developers
filter_data = df[df['Developers'].isin(mylist)]

#Set the properties for the data
flierprops = dict(marker='o', markerfacecolor='r', markersize=5, linestyle='--', markeredgecolor='b')

#Plot the data in the Box plot
boxplot = filter_data.boxplot(column='Price',by='Developers', whis=0.5,notch = True,  flierprops = flierprops,color=dict(boxes='r', medians='b', caps='blue',whiskers ='green'))
boxplot.axes.set_title("Developers vs Price With Notches", fontsize=13)
plt.suptitle('')
boxplot.set_ylabel("Price", fontsize=13)
boxplot = filter_data.boxplot(column='Price',by='Developers',color=dict(boxes='r', medians='b', caps='blue'))
boxprops = dict(linestyle='-', linewidth=4, color='k')
plt.suptitle('')
plt.grid(True)    
boxplot.axes.set_title("Developers vs Price without Notches", fontsize=13)
boxplot.set_xlabel("Developers", fontsize=13)
boxplot.set_ylabel("Price", fontsize=13)
```

# KNN Imputation

KNN imputer is a scikit-learn class used to **predict the missing values in a dataset**. It is the most effectively used method for the numerical data type. 
<br><br>
If the missing values are of type Categorical or Ordinal, the given data are to be **encoded** and converted into equivalent numerical or binary as that is the compatible format for neural networks or machine learning. To implement the KNN imputation, K-Nearest-neighbor values are rendered by finding the neighboring points through a measure of distance(Euclidean Distance).
<br><br>
There are several algorithms available to estimate the neighboring points like Manhattan, Minkowski,Hamming distance, etc among all the methods **Euclidean** is the most common method to find the query points. In the below imputation, numerical data type is chosen with 5 neighboring points taken and estimated with missing values.

```python
#Taking subset of 2000 rows
games_sub = games_df[0:10000]

#Reading csv file and assigning to variable games_impute
games_impute = pd.read_csv('steam_games.csv')
games_impute = games_impute[0:2000]
games_impute.head()
games_impute
```

```python
games_impute['Required age'].unique()
```

```python
games_impute.isna().sum()
```

```python
# import the KNNimputer class
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)

games_impute[['Required age']]= imputer.fit_transform(games_impute[['Required age']])

games_impute[['Required age']].isnull().sum()
```
